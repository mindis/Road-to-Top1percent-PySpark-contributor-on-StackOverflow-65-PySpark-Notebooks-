```
data=[['010', 'DSL', 20191201235900, 'VDRAC'],
['010', 'DSL', 20191203235900, 'VDRAC'],
['010', 'DSL', 20191205235900, 'VDRAC'],
['010', 'DSL', 20200120235900, 'VDRAC'],
['010', 'DSL', 20200128235900, 'VDRAC'],
['010', 'DSL', 20200129235900, 'VDRAC'],
['010', 'DSL', 20200129235900, 'VDRAC']]
cSchema = StructType([StructField("col1", StringType())\
                      ,StructField("col2", StringType())
                      , StructField("col3", LongType())
                      , StructField("col4", StringType())])
df= spark.createDataFrame(data,schema=cSchema)
df2= spark.createDataFrame(data,schema=cSchema)
display(df)
```

```
null=[str(n.nullable) for n in df.schema.fields]
types=[str(i.dataType) for i in df.schema.fields]
both = [list(a) for a in zip(types, null)]
names= df.columns
final = {} 
for key in names: 
    for value in both: 
        final[key] = value
        both.remove(value)
        break
final       
```

```
name=df2 #enter name of dataframe here
def metadata(name):
  null=[str(n.nullable) for n in name.schema.fields]
  types=[str(i.dataType) for i in name.schema.fields]
  both = [list(a) for a in zip(types, null)]
  names= name.columns
  final = {} 
  for key in names: 
     for value in both: 
          final[key] = value
          both.remove(value)
          break
  return final
name.meta= metadata(name)  
```
