```
import pandas as pd
import numpy as np
data_2 = { 'furniture': [np.NaN ,np.NaN ,np.NaN], 'myid': ['1-12', '0-11', '2-12'], 'clothing': ["pants", "shoes", "socks"]} 

df_1 = pd.DataFrame(data_2)
ddf_1 = spark.createDataFrame(df_1)
ddf_1.show() 
```

```
def drop_null_columns(df):
    
    d=df.count()
    null_counts = df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in df.columns]).collect()[0].asDict()
    to_drop = [k for k, v in null_counts.items() if v == d]
    df = df.drop(*to_drop)
    return df
```

```
ddf_2=drop_null_columns(ddf_1)
ddf_2.show()
  
```
